```python
import pandas as pd
data = pd.read_csv("expression_GDSC.csv")
data = data.iloc[:,1:]
data.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE


# Standardize the data (optional but often recommended for t-SNE)
data_standardized = (data - data.mean()) / data.std()

# Apply t-SNE to reduce dimensions
tsne = TSNE(n_components=2, random_state=42)
tsne_components = tsne.fit_transform(data_standardized)
tsne_df = pd.DataFrame(data=tsne_components, columns=['Dimension 1', 'Dimension 2'])
final_df_tsne = pd.concat([tsne_df], axis=1)

# Visualize the reduced dimensions using Seaborn
sns.scatterplot(x='Dimension 1', y='Dimension 2', data=final_df_tsne)
plt.title('t-SNE of expression')
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris  # You can replace this with your dataset

# Load a sample dataset (replace with your dataset)
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
target = pd.Series(data=iris.target, name='species')

# Standardize the data (optional but often recommended for t-SNE)
data_standardized = (data - data.mean()) / data.std()

# Apply t-SNE to reduce dimensions
tsne = TSNE(n_components=2, random_state=42)
tsne_components = tsne.fit_transform(data_standardized)
tsne_df = pd.DataFrame(data=tsne_components, columns=['Dimension 1', 'Dimension 2'])
final_df_tsne = pd.concat([tsne_df, target], axis=1)

# Visualize the reduced dimensions using Seaborn
sns.scatterplot(x='Dimension 1', y='Dimension 2', hue='species', data=final_df_tsne, palette='viridis')
plt.title('t-SNE of Iris Dataset')
plt.show()
```
